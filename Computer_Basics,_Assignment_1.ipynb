1. What is a Computer?

A computer is an electrical device that processes data using a set of
instructions (programs) to execute tasks, solve problems, and produce
results.  It uses both hardware (physical components) and software
(programs/apps) to work. Computer processes data in the machine language.

2. What is RAM?

RAM (Random Access Memory) is a type of volatile memory in a computer that temporarily maintains data and instructions that the CPU (Central Processing Unit) may quickly access while doing tasks.  Unlike hard drives (such as HDDs or SSDs), RAM loses all data when the system is turned off.
